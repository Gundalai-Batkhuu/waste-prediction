```python
# Set the size of the figure
plt.figure(figsize=(8, 8))


# Define the colors for the bars
colors = ['#8B0000', '#0C6606']

ax = sns.countplot('Survived', data=titanic_train, color=colors[0])

plt.title('Survival Rate')
plt.xlabel('Survived')
plt.ylabel('Count')

# Change the color of the bars
for i, p in enumerate(ax.patches):
    if i % 2 == 1:
        p.set_fc(colors[1])

# Add count labels over bars
for p in ax.patches:
    ax.annotate(format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
                
```

```python
# Set the size of the figure
plt.figure(figsize=(8, 8))

sns.violinplot(x='Survived', y='Age', hue='Sex', data=titanic_train)

# Set the title and labels
plt.title('Survival Rate by Age and Gender')
plt.xlabel('Survived')
plt.ylabel('Age')
```

 This plot shows the distribution of ages for survivors and non-survivors, split by gender. We can see that for both male and female passengers, more younger and older people survived.

```python
plt.figure(figsize=(8, 6))
# Create a scatter plot of age and fare, with survival indicated by color
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_train)

# Set the title and labels
plt.title('Age vs. Fare, with Survival Indicated by Color')
plt.xlabel('Age')
plt.ylabel('Fare')
```

 We can see that there is a wide range of fares for all ages, but there are fewer survivors in the lower fare range.

```python
plt.figure(figsize=(7, 6))
# Create a correlation matrix of the numeric variables
corr_matrix = titanic_train[['Survived', 'Age', 'Fare']].corr()

# Create a heatmap of the correlation matrix
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)

# Set the title and labels
plt.title('Correlation Matrix')
# plt.xlabel('Variable')
# plt.ylabel('Variable')
```

There is a **weak positive correlation** between <code>Age</code> and <code>Fare</code>, and a **weak negative correlation** between <code>Age</code> and <code>Survival</code>. The **strongest correlation** is between <code>Fare</code> and <code>Survival</code>, indicating that passengers who paid higher fares were more likely to survive.

```python
sns.factorplot(x ='Survived', y ='Age', data = titanic_train, kind = 'point') 
```

#### **Factor plots shows the probability of survival of passengers by age and sex.** </br>
The <code>x</code> and <code>y</code> parameters define the variables to be plotted, with <code>Survived</code> on the x-axis and <code>Age</code> on the y-axis. The <code>kind</code> parameter is set to <code>point</code> to create a plot with point estimates and confidence intervals for each category.

 In this plot, we can see that the average age of survivors is slightly lower than that of non-survivors. However, the confidence intervals overlap, which suggests that the difference in average age between survivors and non-survivors may not be statistically significant.

```python
sns.factorplot(x ='Survived', y ='Sex', data = titanic_train, kind = 'point')
```

 In the above plot, we can see that a higher proportion of female passengers survived compared to male passengers, which suggests that gender played a role in determining survival in the Titanic disaster.

```python
plt.figure(figsize=(10, 10))
# Create a box plot of fare by passenger class and survival
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=titanic_train)

# Set the title and labels
plt.title('Fare by Passenger Class and Survival')
plt.xlabel('Class')
plt.ylabel('Fare')
```

#### **Key observations:**
- The first and second class passengers, the **median** fare for survivors is higher than that of non-survivors.
- However, for third class passengers, the median fare is similar for both survivors and non-survivors.

```python
plt.figure(figsize=(12, 12))

# Create a pair plot of survival, age, and fare, split by gender
sns.pairplot(data=titanic_train, vars=['Survived', 'Age', 'Fare'], hue='Sex')

# plt.suptitle('Pair Plot of Survival, Age, and Fare, Split by Gender')
```

#### We use pairplot to evaluate **pairwise relationships** between features - here, we are analysing relationships between <code>Survival</code>, <code>Age</code>, and <code>Fare</code>, split by <code>Sex</code>. </br>
- We can see that there is a clear separation between male and female passengers in terms of survival, with female passengers having a higher survival rate. 
- Additionally, we can see that there is a **weak positive correlation** between age and fare for both male and female passengers.

```python
# Create a swarm plot of survival by embarked port and passenger class
sns.swarmplot(x='Embarked', y='Pclass', hue='Survived', data=titanic_train)

# Set the y-axis ticks to show only Pclass 1, 2, and 3
plt.yticks([0, 1, 2, 3], ['0','1', '2', '3'])

# Set the title and labels
plt.title('Survival by Embarked Port and Passenger Class')
plt.xlabel('Embarked Port')
plt.ylabel('Passenger Class')
```

#### The swarmplot is a type of categorical scatterplot that displays the distribution of a numerical variable (y-axis) within categories defined by a categorical variable (x-axis). 

In this, the swarmplot can be used to explore the relationship between survival, the passenger class, and the embarked port.
- Passengers who embarked at Cherbourg generally had a higher survival rate than those who embarked at Queenstown or Southampton.
- Passengers in first class had a higher survival rate than those in second or third class, regardless of where they embarked.
- Passengers in third class had the lowest survival rate, regardless of where they embarked.
